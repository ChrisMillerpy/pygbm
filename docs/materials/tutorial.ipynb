# Simulating Brownian Motions with pygbm

Here you will learn how to simulate Brownian Motions with the pygbm. There are two built in classes that do all the work for you.
- `BMSimulator` - simulates a standard Brownian Motion.
- `GBMSimulator` - simulates a Geometric Brownian Motion.

## Brownian Motion

```python
from pygbm import BMSimulator
```

- First create an instance of the `BMSimulator` class.
    - Give the initial value of your Brownian Motion as $y_0$.

```python
bmsim = BMSimulator(y0=0)
```

- Call the `simulate_path()` method.
    - Give the length of time to simulate the Brownian Motion for as $T$.
    - Give the number of steps that the Brownian Motion will take as $N$.
- This returns two arrays
    - The first is the array of `t_values` that the Brownian Motion is simulated at.
    - The second is the array of `y_values` that the Brownian Motion took at each time $t$.

```python
t_values, y_values = bmsim.simulate_path(T=1, N=1000)
```

- Do with the output what you please, for instance you can plot it with matplotlib

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(5, 5))
fig.suptitle("Brownian Motion Simulation", fontsize=16)
ax.set_ylabel("$B(t)$")
ax.set_xlabel("$t$")
ax.grid(visible=True)
ax.plot(t_values, y_values, linewidth=1)
plt.show();
```

- Now you can get fruity with it.

```python
fig, ax = plt.subplots(figsize=(5, 5))
fig.suptitle("Many Brownian Motions", fontsize=16)
ax.grid(visible=True)
ax.set_ylabel("$B(t)$")
ax.set_xlabel("$t$")

for i in range(10):
    bmsim = BMSimulator(0)
    t_values, y_values = bmsim.simulate_path(1, 100)
    ax.plot(t_values, y_values, linewidth=1)

plt.show();
```
